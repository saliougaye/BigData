# Analizzare i dati degli studenti raccolti la prima lezione

- Importare il csv

```python
# va scaricato lxml e html5lib
import pandas as pd

df = pd.read_csv('./data.csv', sep=',')
df.head()
```

- Contare il numero di maschi e di femmine

```python
df.groupby(['sesso']).size().reset_index(name='Numero')
```

- Contare studenti raggruppati per anno di nascita

```python
df.groupby(['data']).size().reset_index(name='Numero').sort_values(by='Numero', ascending=False)
```

- Contare studenti raggruppati per segno zodiacale

```python
df.groupby(['segnozodiacale']).size().reset_index(name='Numero').sort_values(by='Numero', ascending=False)
```
