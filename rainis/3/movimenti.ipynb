```python
import pandas as pd

df = pd.read_csv('./movimenti.csv', sep=';')

df.head()
```

- Chi sono le 3 persone con saldo maggiore?

```python
df.groupby(['cognome', 'nome']).sum('importo').reset_index().sort_values(by='importo', ascending=False).head(3)
```

- Chi sono le 3 persone con saldo minore?

```python
df.groupby(['cognome', 'nome']).sum('importo').reset_index().sort_values(by='importo', ascending=True).head(3)
```

- Calcolare importo totale movimentato per
ciascun anno

```python
df['data'] = pd.to_datetime(df['data'])

df.groupby(df['data'].dt.year).sum('importo').reset_index()
```

- Elencare le operazioni al di sopra dei 50000 €

```python
gi = df.groupby(['cognome', 'nome']).sum('importo').sort_values(by=['importo'],ascending=False)

gi[gi['importo'] > 50000]

```

- Segnalare le operazioni sospette della persona con il saldo maggiore. Sono quelle il cui importo è superiore a 10 volte l'importo medio.

```python

# troviamo la persona con il saldo maggiore
sm = df.groupby(['cognome', 'nome']).sum('importo').reset_index().sort_values(by='importo', ascending=False).head(1)

# prendiamo tutti i movimenti della persona con saldo maggiore
sm = df[(df['cognome'] == sm.iloc[0]['cognome']) & (df['nome'] == sm.iloc[0]['nome'])]

# calcolo la media dei movimenti
mean = sm['importo'].mean()

# premdi i movimenti con importo 10 volte maggiore
sm[sm['importo'] > mean * 10]

```
